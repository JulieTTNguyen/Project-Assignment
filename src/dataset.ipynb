```python
import pandas as pd
import networkx as nx
```

```python
hyperlinks = pd.read_csv("../data/redditHyperlinks-column-filtered.csv")
wiki = pd.read_csv("../data/wiki_df.csv")
```

```python
# find subreddits represented in the hyperlinks dataset
hyperlink_subreddits = set(hyperlinks['SOURCE_SUBREDDIT'].tolist()+hyperlinks['TARGET_SUBREDDIT'].tolist())
print('Number of subreddits in the hyperlinks dataset:', len(hyperlink_subreddits))

# find subreddits in the wiki dataset
wiki_subreddits = wiki['SUBREDDIT_NAME'].tolist()
print('Number of subreddits in the wiki dataset:', len(wiki_subreddits))

# find subreddits that are in both datasets
subreddits = hyperlink_subreddits.intersection(wiki_subreddits)
print('Number of subreddits in both datasets:', len(subreddits))
```
